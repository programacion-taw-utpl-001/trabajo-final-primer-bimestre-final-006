```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge
from bokeh.core.properties import value
output_notebook()

ciudadess = ['Quito','Guayaquil','Cuenca','Machala','Ambato']



data = {'ciudades' : ciudadess,
       'Quito-1' : [97.1,0,0,0,0],
       'Guayaquil-1' : [149.3,0,0,0,0],
       'Cuenca-1': [93.4,0,0,0,0],
       'Machala-1' : [152.8,0,0,0,0],
       'Ambato-1' : [118.2,0,0,0,0]}

source = ColumnDataSource(data=data)

p = figure(x_range = ciudadess, y_range = (0,200),plot_height=350,title="Pobreza Ecuador - INCIDENCIA - 2008/2017")


p.vbar(x=dodge('ciudades', 0 ,range = p.x_range),top='Quito-1',width=0.20,source=source, color="blue",legend=value("Q"))

p.vbar(x=dodge('ciudades', 1, range = p.x_range),top='Guayaquil-1',width=0.2,source=source, color="red",legend=value("G"))

p.vbar(x=dodge('ciudades', 2, range = p.x_range),top='Cuenca-1',width=0.2,source=source, color="green",legend=value("C"))

p.vbar(x=dodge('ciudades', 3, range = p.x_range),top='Machala-1',width=0.2,source=source, color="yellow",legend=value("M"))

p.vbar(x=dodge('ciudades', 4, range = p.x_range),top='Ambato-1',width=0.2,source=source, color="black",legend=value("A"))



p.x_range.range_padding =0
p.xgrid.grid_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "horizontal"

show(p)
```
