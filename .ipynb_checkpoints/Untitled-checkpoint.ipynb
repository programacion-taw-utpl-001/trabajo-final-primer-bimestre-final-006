```python
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge
from bokeh.core.properties import value
output_notebook()
```

```python
output_notebook()

ciudadess = ['Quito','Cuenca']


data = {'ciudades': ciudadess,
        'Quito-1' : [151,151],
       'Cuenca-1' : [101,145]}

source = ColumnDataSource(data=data)

p = figure(x_range = ciudadess, y_range = (0,200),plot_height=350,title="Pobreza Ecuador", toolbar_location = None, tools="")

p.vbar(x=dodge('ciudades', -0.25,range = p.x_range),top='Quito-1',width=0.2,source=source,
      color="blue",legend=value("Q"))
p.vbar(x=dodge('ciudades', -0.0, range = p.x_range),top='Cuenca-1',width=0.2,source=source,
      color="red",legend=value("C"))

p.x_range.range_padding =0.1
p.xgrid.grid_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
